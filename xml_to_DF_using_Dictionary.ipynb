```python
import pandas as pd
from xml.dom import minidom
```

```python
xml_doc = minidom.parse("C:/Users/User/Desktop/population.xml")
```

```python
xmldata = xml_doc.getElementsByTagName("data")[0]
xmldata
```

```python
records = xmldata.getElementsByTagName("record")
records
```

```python
data = {'Country':[], 'Year':[], 'Population':[]}

for x in records:
    country = x.getElementsByTagName("field")[0].firstChild.data
    year = x.getElementsByTagName("field")[2].firstChild.data
    if x.getElementsByTagName("field")[3].firstChild is None:
        continue
    else:
        population = x.getElementsByTagName("field")[3].firstChild.data
    
    print(country, year, population)      

    data['Country'].append(country)
    data['Year'].append(year)
    data['Population'].append(population)   
    
    

```

```python
df_x = pd.DataFrame(data)
df_indexed = df_x.set_index('Country')
df_indexed
```
